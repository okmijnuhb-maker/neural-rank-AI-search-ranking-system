```python
from datasets import load_dataset

print("Loading MS MARCO dataset...")
dataset = load_dataset("ms_marco", "v1.1", split="train[:5000]")

print(f"Loaded {len(dataset)} queries!")
print("\nSample query:")
print(dataset[0])
```

```python
import json

data_list = []
for item in dataset:
    data_list.append({
        "query_id": item["query_id"],
        "query": item["query"],
        "passages": item["passages"]["passage_text"],
        "labels": item["passages"]["is_selected"]
    })

with open("../data/msmarco_5000.json", "w") as f:
    json.dump(data_list, f)

print(f"Saved {len(data_list)} queries!")
```

```python
print(f"Total queries: {len(data_list)}")
print(f"\nQuery: {data_list[0]['query']}")
print(f"Total passages: {len(data_list[0]['passages'])}")
print(f"Relevance labels: {data_list[0]['labels']}")
print(f"Relevant passage index: {data_list[0]['labels'].index(1)}")
print(f"\nRelevant passage: {data_list[0]['passages'][data_list[0]['labels'].index(1)][:200]}")
```

```python
no_relevant = sum(1 for item in data_list if 1 not in item['labels'])
has_relevant = len(data_list) - no_relevant

print(f"Queries WITH relevant passage: {has_relevant}")
print(f"Queries WITHOUT relevant passage: {no_relevant}")
print(f"Average passages per query: {sum(len(item['passages']) for item in data_list) / len(data_list):.1f}")
```

```python
clean_data = [item for item in data_list if 1 in item['labels']]

with open("../data/clean_msmarco.json", "w") as f:
    json.dump(clean_data, f)

print(f"Clean queries saved: {len(clean_data)}")
```

```python
with open("../data/clean_msmarco.json", "r") as f:
    verify = json.load(f)

print(f"Verified: {len(verify)} clean queries ready!")
```
